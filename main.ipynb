```python
class MyLoss(nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self):
        return
```

```python
LR = 1e-3
EPOCHS = 100

# Create dataloaders for train and test set
train_set = MyDataset()
train_loader = DataLoader(train_set,batch_size=16,shuffle=True)
test_set = MyDataset()
test_loader = DataLoader(train_set)

# Initialize stuff needed for training
model = MyModel()
optimizer = optim.Adam(model.parameters(),LR)
criterion = MyLoss()
plot = loss_plot()

for epoch in range(EPOCHS):
    loss_tr = []
    model.train()
    # One epoch of training
    for inputs, labels in train_loader:
        optimizer.zero_grad()

        preds = model(inputs)
        loss = criterion(preds, labels)
        loss_tr.append(loss.item())

        loss.backward()
        optimizer.step()
    loss_tr = sum(loss_tr)/len(loss_tr)
    plot.plot(loss_tr)

    loss_te = []
    model.eval()
    # Calculate test accuracy for the epoch
    with torch.no_grad():
        for inputs, labels in test_loader:
            preds = model(inputs)
            loss = criterion(preds, labels)
            loss_te.append(loss.item())
    
    loss_te = sum(loss_te)/len(loss_te)

    # Plot train and test loss
    plot.plot(loss_tr,loss_te,LR)
```
